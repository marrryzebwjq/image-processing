```python
def trouver_les_cotes(d, b):
    """
    Retourne la liste des paires (a,c) entières positives telles que
    (d^2 - a^2)*(d^2 - c^2) == (d*b + c*a)^2
    et 0 < a <= c < d (par défaut).
    """
    a_min = 1
    a_max = d-1
    c_min = 1
    c_max = d-1

    sols = []
    for a in range(a_min, min(a_max, d-1)+1):
        for c in range(max(a, c_min), min(c_max, d-1)+1):  # a <= c < d
            left = (d*d - a*a) * (d*d - c*c)
            right = (d*b + c*a)**2
            if left == right:
                sols.append((a, c))
    return sols


for d in range(1, 101):
    for b in range(1, d):
        sols = trouver_les_cotes(d, b)
        if sols:
            print(f"solutions: a={sols[0][0]}, b={b}, c={sols[0][1]}, d={d}")
```
